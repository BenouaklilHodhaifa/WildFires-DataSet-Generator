```python
import numpy as np
import pandas as pd
import requests
from utils import coordinates_selecter
```

```python
coordinates = coordinates_selecter(lat_min=35, lat_max=36, lon_min=2, lon_max=3)
print(coordinates)
```

```python
lat = ''
long = ''
for i in range(len(coordinates)):
    lat += str(coordinates[i][0]) + ','
    long += str(coordinates[i][1]) + ','
payload = {
    'latitude': lat,
    'longitude': long,
    'start_date': '2024-01-01',
    'end_date': '2024-01-02',
    'daily': 'temperature_2m_max'
}
r = requests.get('https://archive-api.open-meteo.com/v1/archive', params=payload)
data = r.json()
```

```python
meteo_data = {
    'time': [],
    'latitude': [],
    'longitude': [],
    'temperature': []
}

for i in range (len(data)):
    res = data[i]
    for j in range(len(res['daily']['time'])):
        meteo_data['time'].append(res['daily']['time'][j])
        meteo_data['latitude'].append(res['latitude'])
        meteo_data['longitude'].append(res['longitude'])
        meteo_data['temperature'].append(res['daily']['temperature_2m_max'][j])

df = pd.DataFrame(meteo_data)
print(df)
```
